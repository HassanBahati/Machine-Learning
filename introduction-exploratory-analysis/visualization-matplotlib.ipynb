```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#the location of data file
filepath = '../data/Iris_Data.csv'

#import the data
data = pd.read_csv(filepath)
```

```python
plt.plot(data.sepal_length,
data.sepal_width, ls='', marker='o')
```

```python
# basic Scatter plots with matplotlib 
# multiple layers of data can also be added 

plt.plot(data.sepal_length,
data.sepal_width,
ls="", marker='o',
label="sepal")

plt.plot(data.petal_length, data.petal_width, ls='', marker='o', label='petal')
```

```python
# histograms can be created from pandas series 
plt.hist(data.sepal_length, bins=25)
```

```python
# customizing matplotlib 
fig, ax = plt.subplot()

ax.barh(np.arange(10), data.sepal_width.iloc[:10])

#set position of ticks and tick labels
ax.set_yticks(np.arrange(0.4, 10, 1.0))
ax.set_ytickLavels(np.arrange(1,11))
ax.set(xlabel="xlabel", ylabel="ylabel", title="Title")
```

```python
# statistial calculations can be included with pandas methods 

data.groupby('species').mean().plot(color=["red","blue","black","green"], fontsize=10.0, figsize=(4,4))
```
